```python
import os
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import keras
from keras.applications import mobilenet_v2
from keras.preprocessing import image

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, ENGLISH_STOP_WORDS
from nltk.tokenize import TreebankWordTokenizer
from nltk.stem import PorterStemmer
from sklearn.decomposition import LatentDirichletAllocation, NMF
```

```python
def prep_image(img_path):

    img = image.load_img(img_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = mobilenet_v2.preprocess_input(x)
    return x
```

```python
model = mobilenet_v2.MobileNetV2(weights='imagenet')

x = prep_image('train_jpg/obj1655_frame0000054.jpg')

out = model.predict(x)
```

```python
mobilenet_v2.decode_predictions(out)
```

```python
mobilenet_v2.decode_predictions(out)[0][0][1] + " " + mobilenet_v2.decode_predictions(out)[0][1][1]
```

```python
image_files = os.listdir("train_jpg")
val_files = os.listdir("val_jpg")
test_files = os.listdir("test_jpg")

test_files_true = []
for file in test_files:
    if file != '.DS_Store':
        test_files_true.append(file)
```

```python
obj_list = []
bio_list = []

for file in image_files:
    if "obj" in file:
        obj_list.append(file)
    else:
        bio_list.append(file)

obj_images_train = [prep_image("train_jpg/" + file) for file in obj_list]
bio_images_train = [prep_image("train_jpg/" + file) for file in bio_list]
```

```python
obj_list = []
bio_list = []

for file in val_files:
    if "obj" in file:
        obj_list.append(file)
    else:
        bio_list.append(file)

obj_images_val = [prep_image("val_jpg/" + file) for file in obj_list]
bio_images_val = [prep_image("val_jpg/" + file) for file in bio_list]        
```

```python
obj_list = []
bio_list = []

for file in test_files_true:
    if "obj" in file:
        obj_list.append(file)
    else:
        bio_list.append(file)

obj_images_test = [prep_image("test_jpg/" + file) for file in obj_list]
bio_images_test = [prep_image("test_jpg/" + file) for file in bio_list]  
```

```python
keywords_obj = []
keywords_bio = []

def keyword(image_list, keyword_list):
    for i in range(len(image_list)):
        print(i)
        x = model.predict(image_list[i])
        y = mobilenet_v2.decode_predictions(x)
        keyword_list.append(y[0][0][1] + " " + y[0][1][1] + " " + y[0][2][1] + " " + y[0][3][1] + " " + y[0][4][1])
    return keyword_list   

keyword(obj_images_train, keywords_obj)
keyword(obj_images_val, keywords_obj)
keyword(obj_images_test, keywords_obj)

keyword(bio_images_train, keywords_bio)
keyword(bio_images_val, keywords_bio)
keyword(bio_images_test, keywords_bio)
```

```python
keywords_obj = list(set(keywords_obj))
```

```python
keywords_bio = list(set(keywords_bio))
```

```python
tfidf = TfidfVectorizer(stop_words=ENGLISH_STOP_WORDS)

tweet_word_matrix_obj = tfidf.fit_transform(keywords_obj)
vocab_obj = tfidf.get_feature_names()
tweet_word_matrix_bio = tfidf.fit_transform(keywords_bio)
vocab_bio = tfidf.get_feature_names()

nmfobj = NMF(n_components=10)
nmfbio = NMF(n_components=10)

nmfobj.fit(tweet_word_matrix_obj)
nmfbio.fit(tweet_word_matrix_bio)
```

```python
tweet_topic_matrix_obj = nmfobj.transform(tweet_word_matrix_obj)
tweet_topic_matrix_df_obj = pd.DataFrame(tweet_topic_matrix_obj).add_prefix('topic_')

word_topic_matrix_df_obj = pd.DataFrame(nmfobj.components_, columns=vocab_obj).T.add_prefix('topic_')
```

```python
tweet_topic_matrix_bio = nmfbio.transform(tweet_word_matrix_bio)
tweet_topic_matrix_df_bio = pd.DataFrame(tweet_topic_matrix_bio).add_prefix('topic_')

word_topic_matrix_df_bio = pd.DataFrame(nmfbio.components_, columns=vocab_bio).T.add_prefix('topic_')
```

```python
def top_words(word_topic_matrix_df, topic, n_words):
    return (word_topic_matrix_df.sort_values(by=topic, ascending=False).head(n_words))[topic]
```
